# Dekorator mierzenia czasu:

```python
import time
```

```python
def measure_time(func):
    def wrapper(*args, **kwargs):
        start = time.perf_counter()
        result = func(*args, **kwargs)
        end = time.perf_counter()
        
        print(f"{func.__name__} took {end - start:.6f} seconds")
        return result
    return wrapper
```

```python
import random
data = [random.randint(0, 10000) for _ in range(1000)]
print(data)
```

# Bubble Sort:

```python
@measure_time
def bubble_sort(arr):
    arr = arr.copy()

    n = len(arr)
    for i in range(n):
        for j in range(0, n - i - 1):
            if arr[j] > arr[j + 1]:
                arr[j], arr[j + 1] = arr[j + 1], arr[j]
    return arr
```

```python
sorted_data = bubble_sort(data)
print(sorted_data)
```

# Selection Sort:

```python
@measure_time
def selection_sort(arr):
    arr = arr.copy()

    n = len(arr)
    for i in range(n):
        min_idx = i

        for j in range(i + 1, n):
            if arr[j] < arr[min_idx]:
                min_idx = j

        arr[i], arr[min_idx] = arr[min_idx], arr[i]
    return arr
```

```python
sorted_data = selection_sort(data)
print(sorted_data)
```

# Insertion Sort:

```python
@measure_time
def insertion_sort(arr):
    arr = arr.copy()

    n = len(arr)
    for i in range(1, n):
        key = arr[i]
        j = i - 1

        while j >= 0 and arr[j] > key:
            arr[j + 1] = arr[j]
            j -= 1

        arr[j + 1] = key
    return arr
```

```python
sorted_data = insertion_sort(data)
print(sorted_data)
```

# Merge Sort:

```python
#@measure_time <-- This is not used because of the recursive nature of merge_sort
def merge_sort(arr):
    arr = arr.copy()

    if len(arr) <= 1:
        return arr
    
    mid = len(arr) // 2
    left = merge_sort(arr[:mid])
    right = merge_sort(arr[mid:])

    return merge(left, right)

def merge(left, right):
    result =[]
    i = j = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            result.append(left[i])
            i += 1
        else:
            result.append(right[j])
            j += 1
        
    result.extend(left[i:])
    result.extend(right[j:])
        
    return result
```

```python
start = time.perf_counter()
sorted_data = merge_sort(data)
end = time.perf_counter()
print(f"merge_sort took {end - start:.6f} seconds")
print(sorted_data)
```

# Quick Sort:

```python
#@measure_time <-- This is not used because of the recursive nature of quick_sort
def quick_sort(arr):
    arr = arr.copy()

    if len(arr) <= 1:
        return arr
    
    pivot = arr[0]

    left = [x for x in arr[1:] if x < pivot]
    right = [x for x in arr[1:] if x >= pivot]

    return quick_sort(left) + [pivot] + quick_sort(right)
```

```python
start = time.perf_counter()
sorted_data = quick_sort(data)
end = time.perf_counter()
print(f"quick_sort took {end - start:.6f} seconds")
print(sorted_data)
```

# Heap Sort:

```python
@measure_time
def heap_sort(arr):
    arr = arr.copy()
    n = len(arr)

    for i in range(n // 2 -1, -1, -1):
        heapify(arr, n , i)

    for i in range(n-1, 0, -1):
        arr[i], arr[0] = arr[0], arr[i]
        heapify(arr, i, 0)

    return arr

def heapify(arr, n, i):
    largest = i
    left = 2 * i + 1
    right = 2 * i + 2

    if left < n and arr[left] > arr[largest]:
        largest = left

    if right < n and arr[right] > arr[largest]:
        largest = right

    if largest != i:
        arr[i], arr[largest] = arr[largest], arr[i]
        heapify(arr, n, largest)
```

```python
sorted_data = heap_sort(data)
print(sorted_data)
```

# Wyszukiwanie binarne

```python
def wyszukiwanie_binarne(tablica, x):
    lewy = 0
    prawy = len(tablica) - 1

    while lewy <= prawy:
        srodek = (lewy + prawy) // 2

        if tablica[srodek] == x:
            return srodek
        if tablica[srodek] < x:
            lewy = srodek + 1
        else:
            prawy = srodek - 1
    
    return -1
```

```python
print(wyszukiwanie_binarne(sorted_data, 118))
```
