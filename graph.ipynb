# DFS - Depth First Search - LIFO

```python
graph = {
    "A": ["B", "C"],
    "B": ["D"],
    "C": [],
    "D": []
}

visited = set()

def dfs(node):
    if node in visited:
        return
    
    visited.add(node)
    print(node)

    for neighbor in graph[node]:
        dfs(neighbor)

dfs("A")
```

# BFS - Breadth First Search - FIFO

```python
from collections import deque

graph = {
    "A": ["B", "C"],
    "B": ["D"],
    "C": [],
    "D": []
}

visited = set()
queue = deque(["A"])

while queue:
    node = queue.popleft()

    if node in visited:
        continue

    visited.add(node)
    print(node)

    for neighbor in graph[node]:
        queue.append(neighbor)
```

# Algorytm Dijkstry

```python
graph = {
    "Warszawa": [("Lodz", 130), ("Gdansk", 340), ("Lublin", 170)],
    "Lodz": [("Warszawa", 130), ("Poznan", 210), ("Wroclaw", 220)],
    "Poznan": [("Lodz", 210), ("Szczecin", 270), ("Wroclaw", 180)],
    "Wroclaw": [("Lodz", 220), ("Poznan", 180), ("Krakow", 270)],
    "Krakow": [("Wroclaw", 270), ("Lublin", 300), ("Rzeszow", 160)],
    "Lublin": [("Warszawa", 170), ("Krakow", 300), ("Rzeszow", 170)],
    "Gdansk": [("Warszawa", 340), ("Szczecin", 350)],
    "Szczecin": [("Gdansk", 350), ("Poznan", 270)],
    "Rzeszow": [("Krakow", 160), ("Lublin", 170)]
}
```

```python
import heapq

def dijkstra(graph, start, end):
    queue = [(0, start)]
    distances = {node: float('inf') for node in graph}
    distances[start] = 0

    previous = {}

    while queue:
        current_distance, current_city = heapq.heappop(queue)
        if current_distance == end:
            break

        for neighbor, weight in graph[current_city]:
            distance = current_distance + weight

            if distance < distances[neighbor]:
                distances[neighbor] = distance
                previous[neighbor] = current_city
                heapq.heappush(queue, (distance, neighbor))

    path = []
    city = end
    while city:
        path.append(city)
        city = previous.get(city)

    path.reverse()

    return distances[end], path

distance, path = dijkstra(graph, "Warszawa", "Krakow")
print("Koszt: ", distance)
print("Trasa: ", " -> ".join(path))
```

```python
import heapq

def dijkstra(graph, start):
    distances = {node: float('inf') for node in graph}
    distances[start] = 0
    
    priority_queue = [(0, start)]
    
    while priority_queue:
        current_distance, current_node = heapq.heappop(priority_queue)
        
        if current_distance > distances[current_node]:
            continue
            
        for neighbor, weight in graph[current_node].items():
            distance = current_distance + weight
            
            if distance < distances[neighbor]:
                distances[neighbor] = distance
                heapq.heappush(priority_queue, (distance, neighbor))
                
    return distances

graph = {
    'A': {'B': 1, 'C': 4},
    'B': {'A': 1, 'C': 2, 'D': 5},
    'C': {'A': 4, 'B': 2, 'D': 1},
    'D': {'B': 5, 'C': 1}
}

print(dijkstra(graph, 'A'))
```

# Wyszukiwanie najbardziej optymalnej trasy prez wszystkie miasta (permutacja)

```python
from itertools import permutations

graph = {
    'A': {'B': 10, 'C': 15, 'D': 20},
    'B': {'A': 10, 'C': 35, 'D': 25},
    'C': {'A': 15, 'B': 35, 'D': 30},
    'D': {'A': 20, 'B': 25, 'C': 30}
}

def tsp(graph, start, end):
    cities = list(graph.keys())
    cities.remove(start)
    cities.remove(end)

    min_cost = float('inf')
    best_path = None

    for path in permutations(cities):
        current_cost = 0
        current_city = start

        for next_city in path:
            current_cost += graph[current_city][next_city]
            current_city = next_city

        current_cost += graph[current_city][end]

        if current_cost < min_cost:
            min_cost = current_cost
            best_path = (start,) + path + (end,)

    return min_cost, best_path

cost, path = tsp(graph, 'A', 'C')

print("Koszt:", cost)
print("Trasa:", " -> ".join(path))
```
