# Ciąg Fibonacciego

```python
def fibbonacci(n):
    if n <= 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fibbonacci(n-1) + fibbonacci(n-2)
```

```python
print(fibbonacci(5))
```

# Algorytm Euklidesa - NWD - najmniejszy wspólny dzielnik

```python
def nwd(a, b):
    while b:
        a, b = b, a % b
    return a
```

```python
print(nwd(282, 78))
```

```python
from math import gcd
print(gcd(282, 78))

import math
print(math.gcd(282, 78))
```

# NWW - największy wspólny dzielnki

```python
def nww(a, b):
    return a * b // nwd(a, b)
```

```python
print(nww(282, 78))
```

```python
from math import lcm
print(lcm(282, 78))

import math
print(math.lcm(282, 78))
```

# Sito Eratostenesa

zwraca liczby pierwsze od przedziału 2 do n

```python
def czy_pierwsza(n):
    if n == 2:
        return True
    if n % 2 == 0 or n <=1:
        return False
    
    for i in range(3, int(n**(1/2)) +1, 2):
        if n % i == 0:
            return False
    return True

print(czy_pierwsza(291))
```

```python
def sito_eratostenesa(n):
    czy_pierwsza = [True] * (n + 1)

    czy_pierwsza[0] = False
    czy_pierwsza[1] = False

    for liczba in range(2, n  + 1):
        if czy_pierwsza[liczba]:
            for wielokrotnosc in range(liczba * liczba, n + 1, liczba):
                czy_pierwsza[wielokrotnosc] = False

    liczby_pierwsze = []

    for liczba in range(2, n + 1):
        if czy_pierwsza[liczba]:
            liczby_pierwsze.append(liczba)
    
    return liczby_pierwsze
```

```python
print(sito_eratostenesa(30))
```

# Suma cyfr

```python
def suma(n):
    suma = 0
    while n:
        suma += n % 10
        n //= 10
    return suma

print(suma(123))
```

# Silnia

```python
def silnia(n):
    if n <= 1:
        return 1
    return n * silnia(n - 1)
print(f"Silnia z liczby 5 wynosi: {silnia(5)}")
```

```python
import math
print(math.factorial(5))
```

# Liczenie wielomianów

```python
def oblicz_wielomian(n, a, x):
    wynik = 0
    potega = 1
    
    # i przyjmuje wartości od 0 do n włącznie
    for i in range(n + 1):
        wynik = wynik + a[i] * potega
        potega = potega * x
        
    return wynik

# Przykład użycia dla wielomianu: W(x) = 2x^2 + 3x + 1 (stopień n = 2)
# Współczynniki: a0 = 1, a1 = 3, a2 = 2 -> lista [1, 3, 2]
# Obliczamy wartość dla x = 5
print(oblicz_wielomian(2, [1, 3, 2], 5))  # Wynik: 1 + 3*5 + 2*25 = 66
```

# Odwrócenie liczby

```python
def odwroc_liczbe(liczba):
    odwrocona = 0
    
    while liczba > 0:
        cyfra = liczba % 10
        odwrocona = odwrocona * 10 + cyfra
        liczba = liczba // 10
        
    return odwrocona

print(odwroc_liczbe(1234))
```
